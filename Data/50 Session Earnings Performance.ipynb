```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def graph(csv_file):
    df = pd.read_csv(csv_file, header = None)
    df[0] = df[0] - 1000
    df[1] = df[1] - 1000
    plt.plot(df[0])
    plt.plot(df[1])
```

```python
#Bayes Bot VS Honest Monte Carlo
graph('bayes_conte.csv')
plt.title("Bayes Bot vs Honest Monte (Conte Marlo) 50 Session Earnings")
plt.legend(['Bayes', 'Conte'])
```

```python
#Bayes Bot vs Fish Bot
graph('bayes_fish.csv')
plt.title("Bayes Bot vs Fish Bot 50 Session Earnings")
plt.legend(['Bayes', 'Fish'])
```

```python
#Bayes Bot vs Heuristic Bot
graph('bayes_heur.csv')
plt.title('Bayes Bot vs Heuristic Bot 50 Session Earnings')
plt.legend(['Bayes', 'Heuristic'])
```

```python
#Bayes vs GMM Monte Carlo
graph('bayes_monte.csv')
plt.title('Bayes Bot vs GMM Monte Carlo 50 Session Earnings')
plt.legend(['Bayes', 'GMM Monte'])
```

```python
#Conte Marlo vs Fish
graph('conte_fish.csv')
plt.title("Conte Marlo vs Fish 50 Session Earnings")
plt.legend(['Conte','Fish'])
```

```python
#Conte Marlo vs Heuristic
graph('conte_heur.csv')
plt.title('Conte Marlo vs Heuristic Bot 50 Session Earnings')
plt.legend(['Conte', 'Heuristic'])
```

```python
#Heuristic vs Fish
graph('heur_fish.csv')
plt.title('Heuristic Bot vs Fish Bot 50 Session Earnings')
plt.legend(['Heuristic', 'Fish'])
```

```python
#Monte Carlo vs Conte Marlo
graph('monte_conte.csv')
plt.title('Monte Carlo vs Conte Marlo 50 Session Earnings')
plt.legend(['Monte Carlo', 'Conte Marlo'])
```

```python
#Monte Carlo vs Fish
graph('monte_fish.csv')
plt.title('Monte Carlo vs Fish 50 Session Earnings')
plt.legend(['Fish', 'Monte Carlo'])
```

```python
#Monte Carlo vs Fish
graph('monte_heur.csv')
plt.title('Monte Carlo vs Heuristic 50 Session Earnings')
plt.legend(['Monte Carlo', 'Heuristic'])
```
