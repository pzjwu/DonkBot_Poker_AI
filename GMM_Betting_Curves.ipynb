```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

```python
x_min_fold = 0.0
x_max_fold = 100.0

mean = 20.0 
std = 10.0

x = np.linspace(x_min_fold, x_max_fold, 100)

fold_freq = scipy.stats.norm.pdf(x,40,10)
call_freq = scipy.stats.norm.pdf(x, 50, 5)
raise_freq = scipy.stats.norm.pdf(x, 68, 10)

bet_curves = np.column_stack((fold_freq, call_freq, raise_freq))

plt.plot(x,fold_freq, color='coral')
plt.plot(x, call_freq, color = 'blue')
plt.plot(x, raise_freq, color = 'green')
plt.title("Normal Distributions of Betting Curves")
plt.legend(["Fold Probability", "Call Probability", "Raise Probability"])

    
```
